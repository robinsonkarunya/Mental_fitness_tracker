```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
df1 = pd.read_csv('mental-and-substance-use-as-share-of-disease.csv')
df2 = pd.read_csv('prevalence-by-mental-and-substance-use-disorder.csv')
df1.head()
```

```python
df = pd.merge(df1,df2)
df
```

```python
df.isnull().sum() #no null values present
```

```python
df = df.drop('Code',axis=1)
df.head()
```

```python
df = df.set_axis(['Country','Year','mental_fitness','Schizophrenia', 'Bipolar_disorder', 'Eating_disorder','Anxiety','drug_usage','depression','alcohol'], axis='columns')
```

```python
df.head()
```

```python
plt.figure(figsize=(12,6))
data = df.drop('Country',axis=1)
sns.heatmap(data.corr(),annot=True)
plt.plot()
```

```python
sns.pairplot(data)
plt.show()
```

```python
plt.bar('Year','mental_fitness',data=data)
plt.xlabel("Year")
plt.ylabel("Mental Fitness")
plt.title('The Mental Fitness over the yearsxc')
```

```python
from sklearn.preprocessing import LabelEncoder
l=LabelEncoder()
for i in df.columns:
    if df[i].dtype == 'object':
        df[i]=l.fit_transform(df[i])
```

```python
from sklearn.model_selection import train_test_split
X = df.drop('mental_fitness',axis=1)
y = df['mental_fitness']

x_train,x_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=2)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,r2_score,accuracy_score

lr = LinearRegression()
lr.fit(x_train,y_train)
y_pred = lr.predict(x_test)
print(mean_squared_error(y_test,y_pred))
print(r2_score(y_test,y_pred))
```

```python
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor()
rf.fit(x_train,y_train)
y_pred_rf = rf.predict(x_test)
print(mean_squared_error(y_test,y_pred_rf))
print(r2_score(y_test,y_pred_rf))
```
